```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['text.usetex'] = True
mpl.rcParams['text.latex.preamble'] = r'\usepackage{libertine}'
mpl.rc('font', family='serif')
```

```python
def m_opt(derivation_1st: float, derivation_2nd: float, t_switch: float, t1: float):
    a = derivation_1st
    b = 0.5 * derivation_2nd
    c = t_switch
    d = t1
    if b == 0 or d == 0:
        raise ValueError("if b == 0 or d == 0")
    under_root = b ** 2 * c ** 2 - a * b * c * d
    if under_root < 0:
        return 1
    m_opt = -(np.sqrt(under_root) + b * c) / (b * d)
    if np.isnan(m_opt):
        raise ValueError("if np.isnan(m_opt)")
    if m_opt < 0:
        print(a, b, c, d)
    if m_opt > 200:
        return 200
    return m_opt
    # return max(1, int(np.ceil(m_opt)))
```

```python
derivation_1 = [1.0 for i in range(1, 101, 1)]
derivation_2 = [-1.0 / i for i in range(1, 101, 1)]
t_switch = [0.1 for i in range(100)]
t1 = [1.0 for i in range(100)]

m = [m_opt(derivation_1st=derivation_1[i], derivation_2nd=derivation_2[i],
           t_switch=t_switch[i], t1=t1[i]) for i in range(100)]

plt.plot(derivation_2, m)
plt.xlabel("2nd derivation")
plt.ylabel("m opt")
plt.show()
```
